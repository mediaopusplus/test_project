```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('新人スタッフ向け問題.xlsx')
df.head()
```

```python
df_before=df.iloc[:,1:5]
df_before
```

```python
df_ymd=df_before[df_before['Unnamed: 1'] == '山田']
df_ymst=df_before[df_before['Unnamed: 1'] == '山下']
df_ymmt=df_before[df_before['Unnamed: 1'] == '山本']
df_iwhn=df_before[df_before['Unnamed: 1'] == '岩花']
df_kwkm=df_before[df_before['Unnamed: 1'] == '川上']
df_ymd
```

```python
def listmake(d):
    l=[]
    for i in range(5):
        if d.iloc[i,2]=='◯':
            l.append(1)
        else:
            l.append(0)
    return l

l_ymd=listmake(df_ymd)
print(l_ymd)
```

```python
df_after = pd.DataFrame(data=[listmake(df_ymd),listmake(df_ymst),listmake(df_ymmt),listmake(df_iwhn),
                              listmake(df_kwkm)], index=['山田','山下','山本','岩花','川上'], 
                        columns=['問題1', '問題2', '問題3','問題4','問題5'])
df_after
```

```python
def scoremake(d):
    s=0
    for i in range(5):
        if(d.iloc[i,2]=='◯'):                                   
            s=s+d.iloc[i,3]
    return s


print(scoremake(df_ymd))
```

```python
def ratemake(d):
    n=0
    for i in range(5):
            if(d.iloc[i,2]=='◯'):
                n=n+1
    r=n/5
    return r       
print(ratemake(df_ymd))
```

```python
df_extra=pd.DataFrame(data=[[scoremake(df_ymd),ratemake(df_ymd)],[scoremake(df_ymst),ratemake(df_ymst)],[scoremake(df_ymmt),ratemake(df_ymmt)],[scoremake(df_iwhn),ratemake(df_iwhn)],[scoremake(df_kwkm),ratemake(df_kwkm)]], 
             index=['山田','山下','山本','岩花','川上'], columns=['得点','正答率'])
df_extra
```

```python
pd.concat([df_after, df_extra], axis=1)
```
